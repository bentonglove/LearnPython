### 善用搜尋工具（關鍵字 for loop）


````python
L = [2,8,3,50]
for i in range(4):
    for j in range(i,4):
        if L[i] > L[j] :
            temp = L[j]
            L[j] = L[i]
            L[i]=temp
            
print(L)
````

## 11.請修改習題5的程式碼：若三數可形成三角形，請判斷圍成的三角形是為銳角、直角或鈍角三角形，然後輸出結果。

例：  
輸入三個邊：  *5 7 6*  
答案： 這是銳角三角形  

輸入三個邊：  *5 3 4*  
答案： 這是直角三角形  



### Tips
1. 對使用者輸入的資料要先做處理
 - 怎樣把輸入的字串切成三個整數？
2. 三個邊的邊長要不要先排序過？
3. 是否合法三角形
 - 三角形三邊合法的判斷式？
4. 上網找三角形 銳角、直角或鈍角 的判斷式？

5. if...then
6. 輸出答案

````python
num_str = input("請輸入三個數字(以半形空白作區隔): ")
num = num_str.split(" ") # 將每個數字分開
# 排序
num.sort()
# 型別轉換
num[0] = int(num[0])
num[1] = int(num[1])
num[2] = int(num[2])
if (num[0]+num[1] > num[2]): # 兩短邊的和大於最長邊
        if (num[0]**2+num[1]**2 == num[2]**2): # 兩短邊的平方和等於最長邊之平方
                print("直角三角形")
        elif (num[0]**2+num[1]**2 < num[2]**2): # 兩短邊的平方和小於最長邊之平方
                print("鈍角三角形")
        else:
                print("銳角三角形")
else:
        print("False")
````

## 12. 金字塔是世界七大奇景之一，請利用for迴圈寫出如下圖所示的4層金字塔圖形。

```python
   * 

  * * 

 * * * 

* * * * 
```


### Tips:
1. for loop
2. 每一層分兩次印, 第一次印空白, 第二次印＊號
3. 找出演算法口訣
4. 使用print 函數的進階功能
### 5.如果印 100層, 要怎寫？

````python
for i in range(4):
    #印空白
    for x in range(3-i):
        print (" ", end="")
    #印 * 號
    for y in range(i+1):
        print ("* ",end="")
        
    print("\n")
````
